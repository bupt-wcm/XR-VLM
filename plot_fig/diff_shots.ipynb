```python
import matplotlib.pyplot as plt
import numpy as np

# 数据准备
x = [1, 2, 4, 8, 12, 16]
y = [60, 65, 70, 75, 80, 85]

# 绘制图表
plt.figure(figsize=(8, 6))  # 设置图表的大小

# 绘制折线
plt.plot([0,1,2,3,4,5], y, label='sin(x)', color='b', linestyle='-', linewidth=2, marker='x', markersize=12, markeredgewidth=2)

plt.plot([0,1,2,3,4,5], y*0.8+10, label='sin(x)', color='b', linestyle='-', linewidth=2, marker='x', markersize=12, markeredgewidth=2)

plt.legend()

# 添加标题
plt.title('Performance of Different Shots on Market1501', fontsize=18, fontweight='bold')

# 添加轴标签
plt.xlabel('Shot Number', fontsize=18)
plt.ylabel('Accuracy', fontsize=18)

# 设置坐标轴刻度
plt.xticks(ticks=[0,1,2,3,4,5], labels=[1, 2, 4, 8, 12, 16], fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(60, 90)

# 添加网格
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# 设置图例
plt.legend(fontsize=12, loc='upper right')

# 设置边框样式
plt.gca().spines['top'].set_linewidth(0.5)
plt.gca().spines['right'].set_linewidth(0.5)

# 展示图表
plt.tight_layout()  # 调整图表布局，防止标签被遮挡
plt.show()
```
